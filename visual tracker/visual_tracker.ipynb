```python
import numpy as np
import cv2

cap=cv2.VideoCapture(0)
#uses first camera

while (1):
    _,frame=cap.read()

    #below is variables of each color, but not only that color
    red=np.matrix(frame[:,:,2]) #B G R
    green=np.matrix(frame[:,:,1])
    blue=np.matrix(frame[:,:,0])

    #separated red image layer
    red_only=np.int16(red)-np.int16(green)-np.int16(blue)
    
    #limits matrix to standard color range
    red_only[red_only<50]=0
    red_only[red_only>=60]=255
    #alter value in brackets to adjust noise in image

    #sets matrix back to image quality
    red_only=np.uint8(red_only)
    

    #mask
    mask=np.ones((5,5),np.uint8)
    opening=cv2.morphologyEx(red_only, cv2.MORPH_OPEN,mask)
    cv2.imshow('Masked Image',opening)

    #begin adding mask
    _,labels=cv2.connectedComponents(opening,4)
    #cuts the mask into number of visible objects
    b=np.matrix(labels)

    #inverts any 0 into a 1 from b
    Obj1=b==0
    #converts variable into image quality
    Obj1=np.uint8(Obj1)
    #increases brightness of image
    Obj1[Obj1>0]=255
    cv2.imshow('First Object',Obj1)

    #inverts any 1 into a 1 from b
    Obj2=b==1
    #converts variable into image quality
    Obj2=np.uint8(Obj2)
    #increases brightness of image
    Obj2[Obj2>0]=255
    cv2.imshow('Second Object',Obj2)

    #inverts any 2 into a 1 from b
    Obj3=b==2
    #converts variable into image quality
    Obj3=np.uint8(Obj3)
    #increases brightness of image
    Obj3[Obj3>0]=255
    cv2.imshow('Third Object',Obj3)

    #inverts any 3 into a 1 from b
    Obj4=b==3
    #converts variable into image quality
    Obj4=np.uint8(Obj4)
    #increases brightness of image
    Obj4[Obj4>0]=255
    cv2.imshow('Fourth Object',Obj4)

    #displays windows
    cv2.imshow('Frame', frame)
    cv2.imshow('Red Only', red_only)



    k=cv2.waitKey(5)
    #will wait until key press (5>?)
    if k==27:
        #breaks on ESC
        break

cv2.destroyAllWindows()
```
